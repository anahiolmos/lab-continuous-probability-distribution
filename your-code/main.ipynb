## Uniform Distribution

To generate uniform random numbers between any two given values using scipy, we can either use the following code or the code that we have
discussed in class:

```python
from scipy.stats import uniform

x = uniform.rvs(size=10)
a = 2
b = 3
randoms  = a + (b-a)*x
print(randoms)
```

**Your task:**

1. Based on the code above, write a function that generates uniformly distributed random numbers. There are several requirements for your function:
    * It should accept 3 parameters: 
        * `bottom` - the lower boundary of the generated numbers
        * `ceiling` - the upper boundary of the generated numbers
        * `count` - how many numbers to generate
    * It should return an array of uniformly distributed random numbers

1. Call your function with 2 sets of params below:
    * bottom=10, ceiling=15, count=100
    * bottom=10, ceiling=60, count=1,000

1. Plot the uniform distributions generated above using histograms, where x axis is the value and y axis is the count. Let the histogram's number of bins be 10.

Your output should look like below:

![uniform distribution](ud.png)

```python
# your code here

def uniformly_random_numbers(bottom, ceiling, count):
    return uniform.rvs(size=count, loc=bottom, scale =(ceiling-bottom))
```

```python
uniformly_random_numbers(10, 15, 100)
```

```python
uniformly_random_numbers(10, 60, 1000)
```

```python
fig, axis = plt.subplots(1, 2, sharey=True)

axis[0].hist(uniformly_random_numbers(bottom=10, ceiling=15, count=100))

axis[1].hist(uniformly_random_numbers(bottom=10, ceiling=60, count=1000));
```

How are the two distributions different?

```python
# your answer below

# even the bottom is the same for both, the ceiling is different and it can be notorious
# for the first one is 15 and for the second is 60, so is much higher the distribution

# also the distribution is greater in the sample of 1000 than in a sample of 100
```

## Normal Distribution

1. In the same way in the Uniform Distribution challenge, write a function that generates normally distributed random numbers.
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 1
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 50
2. Plot the distributions of the data generated.

Expected output:

![normal distribution](nd.png)

```python
# your code here

def normal_random_numbers(avg, std, count):
    return norm.rvs(avg, std, count)
```

```python
normal_random_numbers(100, 1, 1000)
```

```python
normal_random_numbers(100, 50, 1000)
```

```python
fig, axis = plt.subplots(1, 2, sharey=True)

axis[0].hist(normal_random_numbers(10, 1, 1000), bins=50)

axis[1].hist(normal_random_numbers(10, 50, 1000),bins=50);
```

How are the two distributions different?

```python
# your answer below

# the sample with a higher standard deviation (50) has in a higher interval distrusted its data
# its data is farer from the average (10)
# in comparison with the sample with a standard deviation of 1 and the average of 10
```
